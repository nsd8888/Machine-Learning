```python
import sklearn.cluster as cluster
kmns=KMeans(n_clusters=5)
kms=kmns.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
```

```python
df['cluster']=kms
```

```python
#lets define ssc and plot elbow plot
sse=[]
for i in range(1,10):
    kmns=KMeans(n_clusters=i)
    kmns.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
    sse.append(kmns.inertia_)
```
